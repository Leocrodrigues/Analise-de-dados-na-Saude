```python
# Criando uma função que retorna os ultimos(mais recentes casos pela ultima data) por país
def latest_by_country(data):
    return data.groupby('Country/Region').sum().iloc[:,-1]  # fazendo com iloc todas as linhas , a ultima coluna

latest_by_country(confirmed)
```

```python
# Atribuindo os dados confirmados, mortes e recuperados a informação. Lembrando que é do último os dados
information = [latest_by_country(confirmed), latest_by_country(deaths), latest_by_country(recovered)]
```

```python
# concatenando as informações 
combined = pd.concat(information, axis =1)
combined
```

```python
# ALterando o nome das colunas dos dados combinados
combined.columns = ["confirmed", "deaths", "recovered"]
combined.head()
```

```python
# Criando uma função para buscar o total de casos por data no Brasil
def latest_by_country_at(data, date):
    return data.groupby('Country/Region').sum()[date]  


# Atribuindo os dados confirmados, mortes e recuperados a informação. Lembrando que é do último os dados
information = [latest_by_country_at(confirmed, '9/21/21'), latest_by_country_at(deaths, '9/21/21'), latest_by_country_at(recovered, '9/21/21')]

# concatenando as informações 
combined_9_21_21 = pd.concat(information, axis =1)

# ALterando o nome das colunas dos dados combinados
combined_9_21_21.columns = ["confirmed", "deaths", "recovered"]


# Taxa de letalidade  é expressa em porcentagem 
sum_up = combined_9_21_21.loc['Brazil']
letality_rate_1 = sum_up['deaths'] / sum_up['confirmed'] 
letality_rate_2 = sum_up['deaths'] / (sum_up['deaths'] + sum_up['recovered'])
print(letality_rate_1*100)
print(letality_rate_2)
```

```python
# Pegando a taxa de letalidade correta, com 12 dias antes da morte no Brasil
# Atribuindo os dados confirmados, mortes e recuperados a informação. Lembrando que é do último os dados
information = [latest_by_country_at(confirmed, '9/9/21'), latest_by_country_at(deaths, '9/21/21'), latest_by_country_at(recovered, '9/21/21')]

# concatenando as informações 
combined_12 = pd.concat(information, axis =1)

# ALterando o nome das colunas dos dados combinados
combined_12.columns = ["confirmed", "deaths", "recovered"]


# Taxa de letalidade  é expressa em porcentagem 
sum_up = combined_12.loc['Brazil']
letality_rate_3 = sum_up['deaths'] / sum_up['confirmed'] 
print(letality_rate_3*100)
```

### Lendo nova base de dados da população mundial e variantes

https://population.un.org/wpp/Download/Standard/CSV/

```python
# Vamos incluir uma nova base atualizada da população mundial, para descobrir a taxa de incidência de cada país/regiao
un_populacao = pd.read_csv("https://population.un.org/wpp/Download/Files/1_Indicators%20(Standard)/CSV_FILES/WPP2019_TotalPopulationBySex.csv")
un_populacao['PopTotal'] = un_populacao['PopTotal'] * 1000 # definindo a popução em milhares
un_populacao.head()
```

```python
un_populacao.describe()
```

```python
# Pgeando os dados somente de 2021, duas colunas e definindo como indice localização(location)
un_populacao_total = un_populacao.query('Time == 2021')[['Location', 'PopTotal']].set_index('Location')
un_populacao_total.head()
```

```python
# juntando a base de dados combined com da com da pouplaçao tota(un_populacao_total)
combined.join(un_populacao_total)
```

```python
combined.join(un_populacao_total).sort_values("confirmed", ascending=False).head(20)
```

```python
# buscando somente do brasil, utilizando engine do python pois estamos utilizando str.contains
un_populacao.query("Location.str.contains('Kong') and Time == 2021", engine='python')
```

```python
# assumindo algumas premissas
# de que China é  igual Mainland China 
# criando a função
# criando nova coluna da localização segundoa organização(who)
# e renomeando os paises cujos estao com nan
def rename_location(location):
    if location=='China':
        return 'Mainland China'
    if location=='Republic of Korea':
        return 'Korea'
    if location=='Iran (Islamic Republic of)':
        return 'Iran'
    if location=='United States of America':
        return 'US'
    if location=='China, Hong Kong SAR':
        return 'Hong Kong'
    return location 


un_populacao_total['location_da_who'] = un_populacao_total.index.map(rename_location)
un_populacao_total.query("location_da_who  == 'Mainland China'")
```

```python
un_populacao_total = un_populacao_total.set_index('location_da_who')
un_populacao_total.head()
```

```python
un_populacao_total
```

```python
# Agora fazendo novamente o join com os nomes renomeados e dropando o na
combined_expandidos = combined.join(un_populacao_total).sort_values('confirmed', ascending=False).dropna()
combined_expandidos.head(10)
```

```python
# O calculo de incidência aqui, esta errado, pois assumi que a população em risco de Mainland China é todo mundo
# mas até agora 83% dos casos estao em um estado/provincia Hubei
# Portanto esta incorreto considerar esses números 

# Encontrando a taxa incidentes  que é os casos confirmados / pela população total de casa país * por 100mil habitantes
# E tambem a taxa de mortalidade
combined_expandidos['incidence_ratio'] = combined_expandidos['confirmed'] / combined_expandidos['PopTotal'] * 100000
combined_expandidos['mortality_rate'] = combined_expandidos['deaths'] / combined_expandidos['PopTotal'] * 100000

combined_expandidos.head(10)
```
